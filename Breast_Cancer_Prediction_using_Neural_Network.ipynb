```python
#importing dependencies
import torch
import torch.nn as nn
import torch.optim as optim #for optimizer (adam,sgd)
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
#Device Configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)

```

```python
#data collection and pre processing
data = load_breast_cancer()
X = data.data
Y = data.target
```

```python
print(X)
```

```python
print(Y)
```

```python
#split the data set into traning and test
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(X.shape)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
#data standardiastion
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
type(X_train)
```

```python
#we cant use num py in torch so we have to convert it to PyTorch tensars

X_train = torch.tensor(X_train, dtype=torch.float32).to(device)
Y_train = torch.tensor(Y_train, dtype=torch.float32).to(device)
X_test = torch.tensor(X_test, dtype=torch.float32).to(device)
Y_test = torch.tensor(Y_test, dtype=torch.float32).to(device)
```

```python
#neural network architecture in PyTorch

class NeuralNet(nn.Module):

  def __init__(self, input_size, hidden_size, num_classes):
    super(NeuralNet, self).__init__()
    self.fc1 = nn.Linear(input_size, hidden_size)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(hidden_size, num_classes) # Changed output_size to num_classes, assuming num_classes is the intended output size
    self.sigmoid = nn.Sigmoid()

  def forward(self, x):
    out = self.fc1(x)
    out = self.relu(out)
    out = self.fc2(out)
    out = self.sigmoid(out)
    return out
```

```python
#define hyper parameters
input_size = X_train.shape[1]#30
hidden_size = 64
output_size = 1
learning_rate=0.001
num_epochs = 100
```

```python
#initialize the nn and move ot gpu
model = NeuralNet(input_size, hidden_size, output_size).to(device)
```

```python
#define loss and the optimizers
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
#traning the nn
for epoch in range(num_epochs):
  model.train()
  optimizer.zero_grad()
  outputs = model(X_train)
  loss = criterion(outputs, Y_train.view(-1,1))
  loss.backward()
  optimizer.step()

  #calculation accuracy
  with torch.no_grad():
    outputs = model(X_test)
    predicted = outputs.round()
    # Compare with Y_test, not Y_train, and use .float() for casting
    correct = (predicted == Y_test.view(-1,1)).float().sum()
    accuracy = correct / Y_test.size(0)

  # Move print statement inside the loop
  if (epoch+1) % 10 == 0:
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Accuracy: {accuracy.item()*100:.2f}%")
```

```python
#model evaluation
model.eval()
with torch.no_grad():
  outputs = model(X_test)
  predicted = outputs.round()
  correct = (predicted == Y_test.view(-1,1)).float().sum()
  accuracy = correct / Y_test.size(0)
  print(f"accuracy on traning data:{accuracy.item()*100:.2f}%")
```
